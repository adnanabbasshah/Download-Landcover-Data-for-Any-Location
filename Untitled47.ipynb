```python
import rasterio
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap, BoundaryNorm
from matplotlib.patches import Patch

tif_path = '/content/denmark.tif'

class_names = [
    'Water', 'Trees', 'Grass', 'Flooded Vegetation', 'Crops',
    'Shrub And Scrub', 'Built', 'Bare', 'Snow And Ice'
]

palette = [
    '#419bdf', '#397d49', '#88b053', '#7a87c6', '#e49635',
    '#dfc35a', '#c4281b', '#a59b8f', '#b39fe1'
]

cmap = ListedColormap(palette)
norm = BoundaryNorm(np.arange(-0.5, len(class_names)), cmap.N)

with rasterio.open(tif_path) as src:
    img = src.read(1)
    transform = src.transform
    width = src.width
    height = src.height

fig, ax = plt.subplots(figsize=(10, 10))
im = ax.imshow(img, cmap=cmap, norm=norm)

legend_handles = [Patch(color=palette[i], label=class_names[i]) for i in range(len(class_names))]

# legend styling
legend = ax.legend(
    handles=legend_handles,
    loc='upper center',
    bbox_to_anchor=(0.5, -0.15),
    ncol=3,
    frameon=True,
    fontsize=12,
    title='Classes',
    title_fontsize=14,
    borderpad=1,
    labelspacing=0.8,
    columnspacing=1.5
)

# legend frame style
frame = legend.get_frame()
frame.set_edgecolor('gray')
frame.set_linewidth(1)
frame.set_alpha(0.9)
frame.set_boxstyle('round,pad=0.5')

num_ticks = 5
x_ticks_img = np.linspace(0, width, num_ticks)
y_ticks_img = np.linspace(0, height, num_ticks)

x_ticks_geo = [transform * (x, 0) for x in x_ticks_img]
y_ticks_geo = [transform * (0, y) for y in y_ticks_img]

x_ticks_geo_vals = [coord[0] for coord in x_ticks_geo]
y_ticks_geo_vals = [coord[1] for coord in y_ticks_geo]

ax.set_xticks(x_ticks_img)
ax.set_xticklabels([f"{x:.2f}" for x in x_ticks_geo_vals])
ax.set_yticks(y_ticks_img)
ax.set_yticklabels([f"{y:.2f}" for y in y_ticks_geo_vals])

ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
ax.set_title('Land Cover Classification', fontsize=16, fontweight='bold')

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.tight_layout()
plt.show()
```

```python
import rasterio
import geopandas as gpd
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap, BoundaryNorm
from matplotlib.patches import Patch
import numpy as np

# Paths
tif_path = '/content/7d0e2a0ae94ae8d5c6a49ad3e8f79c96-490cdbf6dcc62fd0b3a01094e38bce5c-getPixels.tif'
shp_path = '/content/Islamabad_clip.shp'

# Load raster
with rasterio.open(tif_path) as src:
    img = src.read(1)
    raster_crs = src.crs
    bounds = src.bounds

# Load shapefile
gdf = gpd.read_file(shp_path)

# Reproject shapefile to raster CRS if needed
if gdf.crs != raster_crs:
    gdf = gdf.to_crs(raster_crs)

# Color map setup
class_names = [
    'Water', 'Trees', 'Grass', 'Flooded Vegetation', 'Crops',
    'Shrub And Scrub', 'Built', 'Bare', 'Snow And Ice'
]
palette = [
    '#419bdf', '#397d49', '#88b053', '#7a87c6', '#e49635',
    '#dfc35a', '#c4281b', '#a59b8f', '#b39fe1'
]
cmap = ListedColormap(palette)
norm = BoundaryNorm(np.arange(-0.5, len(class_names)), cmap.N)

# Plotting
fig, ax = plt.subplots(figsize=(10, 10))

# Plot shapefile boundary first
gdf.boundary.plot(ax=ax, edgecolor='red', linewidth=2)

# Plot raster on top with spatial extent and origin set to 'upper' (try 'lower' if upside-down)
extent = (bounds.left, bounds.right, bounds.bottom, bounds.top)
im = ax.imshow(img, cmap=cmap, norm=norm, extent=extent, origin='upper')

# Set axis limits to raster bounds for correct alignment
ax.set_xlim(bounds.left, bounds.right)
ax.set_ylim(bounds.bottom, bounds.top)

# Legend
legend_handles = [Patch(color=palette[i], label=class_names[i]) for i in range(len(class_names))]
ax.legend(handles=legend_handles, loc='upper center', bbox_to_anchor=(0.5, -0.15), ncol=3,
          frameon=True, fontsize=12, title='Classes', title_fontsize=14,
          borderpad=1, labelspacing=0.8, columnspacing=1.5)

# Labels and title
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
ax.set_title('Islamabad Boundary Overlay on Land Cover Classification')

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.tight_layout()
plt.show()
```
